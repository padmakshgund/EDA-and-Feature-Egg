```python
df1 = data.groupby(["Aggregate rating","Rating color","Rating text"]).size().reset_index().rename(columns = {0:"Rating count"} )
df1
```

#### Obeservaion
1. Maximum count is 0 rating
2. rating is between 4.5 to 4.9---> Excellent
3.  rating is between 4.0 to 4.5---> very good
4. rating is between 3.5 to 4---> good
5. rating is between 2.5 to 3.4---> Average
6. rating is between 0 to2.5---> poor

```python
import matplotlib as mt
mt.rcParams["figure.figsize"] = (12,8)
sns.barplot(x ="Aggregate rating",y = "Rating count",hue = "Rating color",data = df1,palette=["white","red",'orange','yellow','green',"green"])
```

##### Observation
1. No rating count is very high
2. Maximum no. of rating is between 2.5 to 3.4

```python
## Count plot
sns.countplot(x = "Rating color",data = df1,palette=["blue","red","orange","yellow","green","green"],);
```
